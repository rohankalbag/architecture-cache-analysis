# Analysis of Program Output

```python
import pandas as pd
x = pd.read_csv("output.csv")
display(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [13, 18]
for i in range(len(x)):
    val = x.iloc[i]
    plt.plot(list(val.keys())[1:], list(val.values)[1:], label=list(val.values)[0], marker='s')
plt.title("Plotted Program Results")
plt.xlabel("Stride")
plt.ylabel("Access Time (ns)")
plt.legend()
plt.show()
```
